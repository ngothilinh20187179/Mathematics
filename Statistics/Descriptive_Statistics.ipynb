Measure of Central Tendency 
- mean
- median
- mode

Measure of Variability
- Range
- Variance
- Standard Deviation
- Interquartile Range

Measures of Frequency Distribution
- Histogram
- Boxplot
- Pie chart

```python
import numpy as np
import pandas as pd

arr = [1,2,3,4,5,6,7,8,2,2]

mean = np.mean(arr)
median = np.median(arr)

series = pd.Series(arr)
mode = pd.Series(arr).mode()[0]

print("Mean", mean)
print("Median", median)
print("Series mode", series)
print("1st Mode", mode)
```

```python
arr2 = [1,2,3,4,5,6]
variance = np.var(arr2)
standard_deviation = np.std(arr2)

print("Mean", np.mean(arr2))
print("Variance", variance)
print("Standard Deviation", standard_deviation)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = [12, 15, 14, 10, 18, 20, 22, 20, 10, 14]
min = np.min(data)
max = np.max(data)
median = np.median(data)

print("Min", min)
print("Max", max)
print("Range", max-min)
print("Median", median)

sns.boxplot(data)
plt.title("Boxplot")
plt.show()
```

```python
data_histogram = np.random.randn(100)

plt.hist(data_histogram, bins=30, edgecolor='black')
plt.title('Histogram của Phân phối Chuẩn')
plt.xlabel('Giá trị')
plt.ylabel('Tần suất')
plt.show()
```

```python

labels = ["A", "B", "C", "D"]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
sizes = [25, 30, 45, 15]
explode = (0, 0.05, 0, 0)

plt.pie(x=sizes , labels=labels, colors=colors, explode=explode, autopct='%1.1f%%')
plt.show()
```
