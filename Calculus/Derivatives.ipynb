```python
import matplotlib.pyplot as plt
import numpy as np

# f(x) = x^2
x = np.linspace(-3, 3, 400)
y = x**2

x0 = 1.0
y0 = x0**2
slope = 2 * x0 

# Phng trnh tip tuyn ti x0: y = f'(x0)*(x - x0) + f(x0)
tangent_line = slope * (x - x0) + y0

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = x^2$', color='blue')
plt.plot(x, tangent_line, label=f'Tip tuyn ti x={x0}', color='red', linestyle='--')
plt.scatter([x0], [y0], color='black')
plt.text(x0 + 0.1, y0 - 1, f"o hm = {slope}", fontsize=12)

plt.title('Minh ho o hm f\'(x) = 2x vi f(x) = x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```
