```python
import matplotlib.pyplot as plt

x1 = [1, 2, 3]
x2 = [4, 5, 6]
y  = [19, 26, 33]

w1 = 0
w2 = 0
b = 0
lr = 0.01
epochs = 100
loss_history = []
n = len(y)

for epoch in range(epochs):
    total_loss = 0
    dw1 = 0
    dw2 = 0
    db = 0

    for i in range(n):
        y_pred = w1 * x1[i] + w2 * x2[i] + b
        error = y_pred - y[i]
        total_loss += error**2

        # Partial_Derivatives
        dw1 += 2 * error * x1[i]
        dw2 += 2 * error * x2[i]
        db  += 2 * error

    dw1 /= n
    dw2 /= n
    db  /= n

    w1 -= lr * dw1
    w2 -= lr * dw2
    b  -= lr * db

    avg_loss = total_loss / n
    loss_history.append(avg_loss)

    if epoch % 10 == 0:
        print(f"Epoch {epoch}: Loss = {avg_loss:.4f}, w1 = {w1:.2f}, w2 = {w2:.2f}, b = {b:.2f}")

plt.plot(loss_history)
plt.xlabel("Epoch")
plt.ylabel("Loss (MSE)")
plt.title("Loss giảm dần theo thời gian (Gradient Descent)")
plt.grid(True)
plt.show()
```
