Discrete distributions: Binomial, Poisson, Geometric... 

Continuous distributions: Normal (Gaussian), Exponential, Uniform...

```python
# Binomial Distribution
# Probability of having k successes in n repeated attempts
# Tung đồng xu 10 lần, xác suất ra mặt ngửa là 0.5 -> Xác suất có đúng k lần ra mặt ngửa?
from scipy.stats import binom
import matplotlib.pyplot as plt

n = 10              # số lần tung đồng xu
p = 0.5             # xác suất ra mặt ngửa
k = range(0, 11)    # k=0,1,2,...,10

# P(X = k)
probs = binom.pmf(k, n, p)

plt.bar(k, probs)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Số lần mặt ngửa (k)")
plt.ylabel("Xác suất P(X=k)")
plt.show()
```

```python
# Poisson Distribution
# Average : 3 calls / min → pro k calls / 1 min

from scipy.stats import poisson

mu = 3  # số trung bình (λ)
x = range(0, 10)

probs = poisson.pmf(x, mu)

plt.bar(x, probs)
plt.title("Poisson Distribution (λ=3)")
plt.xlabel("Số cuộc gọi (k)")
plt.ylabel("Xác suất P(X=k)")
plt.show()
```

```python
# Geometric Distribution
# Xác suất trúng thưởng là 0.3 -> Xác suất lần đầu tiên trúng vào lần thứ k?

from scipy.stats import geom

p = 0.3
x = range(1, 11)

probs = geom.pmf(x, p)

plt.bar(x, probs)
plt.title("Geometric Distribution (p=0.3)")
plt.xlabel("Lần đầu trúng (k)")
plt.ylabel("Xác suất P(X=k)")
plt.show()
```

```python
# Normal (Gaussian) Distribution
# Chiều cao trung bình là 170cm, độ lệch chuẩn 10cm
from scipy.stats import norm
import numpy as np

mu = 170
sigma = 10
x = np.linspace(130, 210, 500)
pdf = norm.pdf(x, mu, sigma)

plt.plot(x, pdf)
plt.title("Normal Distribution (μ=170, σ=10)")
plt.xlabel("Chiều cao (cm)")
plt.ylabel("Mật độ xác suất")
plt.grid(True)
plt.show()
```

```python
# Exponential Distribution
# Thời gian giữa 2 lần gọi đến tổng đài, trung bình 5 phút

from scipy.stats import expon

scale = 5  # trung bình = 1/λ
x = np.linspace(0, 30, 500)
pdf = expon.pdf(x, scale=scale)

plt.plot(x, pdf)
plt.title("Exponential Distribution (mean=5 mins)")
plt.xlabel("Thời gian chờ (phút)")
plt.ylabel("Mật độ xác suất")
plt.grid(True)
plt.show()
```

```python
# Uniform Distribution
# Một người có thể đến từ 0 đến 10 phút → thời gian đến là phân phối đều
from scipy.stats import uniform

a = 0
b = 10
x = np.linspace(-2, 12, 500)
pdf = uniform.pdf(x, loc=a, scale=b-a)

plt.plot(x, pdf)
plt.title("Uniform Distribution (0–10 mins)")
plt.xlabel("Thời gian đến (phút)")
plt.ylabel("Mật độ xác suất")
plt.grid(True)
plt.show()
```
